```python
import re
import numpy as np


def modify_params_chem(str_vec, content_vec, params_mod):
    
    idx_interest = str_vec.index(params_mod[0])
    content_vec[idx_interest]['rate_expr'][0] = str(params_mod[1])
    
    return content_vec


params_mod = ['O(3P) + wall -> 0.5O2(X)', 1.4e-3]

def change_chem(in_path, out_path, params_mod):
    
    str_vec = []
    content_vec = []
    with open(in_path) as f:
        
        for raw in f:
            line = raw.rstrip()
            # Skip section header
            m = re.match(r'%[- ]*([^-]+?) *[-]*$', line)
            if m:
                continue
            # Skip blank or pureâ€comment lines
            if not line or line.lstrip().startswith('%'):
                continue
            # Split on the first 3 '|'s
            parts = [p.strip() for p in line.split('|', 3)]
            if len(parts) != 4:
                continue

            reac_prod, rate_type, rate_expr, comment = parts
            rate_expr = rate_expr.split(',')
            if not '%' in comment:
                comment = '%' + comment
                
            str_vec.append(reac_prod)
            content_vec.append({
                'rate_type':  rate_type,
                'rate_expr':  rate_expr,
                'reference':  comment
            })
            
    
    content_vec = modify_params_chem(str_vec, content_vec, params_mod)
    
        
    ###* write the output
    output_txt = []
    for i in range(len(content_vec)):
        content_vec[i]['rate_expr'] = ','.join(content_vec[i]['rate_expr'])
        line2 = '\t| '.join(list(content_vec[i].values()))
        line = str_vec[i] + "\t| " + line2 + "\n"
        output_txt.append(line)
    
    with open(out_path, 'w') as fout:
        fout.writelines(output_txt)
    


# Example usage:
if __name__ == "__main__":
    
    in_path = 'LoKI_V3/Code/Input/Oxygen/oxygen_novib_thermalflux_effective.chem'
    out_path = 'LoKI_V3/Code/Input/Oxygen/oxygen_novib_thermalflux_effective.chem'
    
    params_mod = ['O(3P) + wall -> 0.5O2(X)', 1.4e-3]
    
    change_chem(in_path, out_path, params_mod)





# # Example usage:
# if __name__ == "__main__":
#     str_vec, content_vec = load_chem('LoKI_V3/Code/Input/Oxygen/oxygen_novib_thermalflux_effective.chem')

    
#     ####* modify value

#     params_mod = ['O(3P) + wall -> 0.5O2(X)', 1.4e-3]

#     idx_interest = str_vec.index(params_mod[0])
#     content_vec[idx_interest]['rate_expr'][0] = str(params_mod[1])
    
    
    
#     ###* write the output
#     output_txt = []
#     for i in range(len(content_vec)):
#         content_vec[i]['rate_expr'] = ','.join(content_vec[i]['rate_expr'])
#         line2 = '\t| '.join(list(content_vec[i].values()))
#         line = str_vec[i] + "\t| " + line2 + "\n"
#         output_txt.append(line)
    
    
#     out_file = 'LoKI_V3/Code/Input/Oxygen/oxygen_novib_thermalflux_effective2.chem'
#     with open(out_file, 'w') as fout:
#         fout.writelines(output_txt)
    
    
    
```
