```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python

##* baselines algorithms

with h5py.File("../study_opt_local/results/results_Powell_train_V4.h5", "r") as f:
    best_loss_powell = f['best_loss'][:]
    iters_powell = f['iters'][:]+1


with h5py.File("../study_opt_local/results/results_DE_40_train_V4.h5", "r") as f:
    best_loss_de = f["best_loss"][:]
    iters_de = f['iters'][:]+1


with h5py.File("../study_opt_local/results/results_CMA_ES_train_V4.hdf5", "r") as f:
    best_loss_cma = f["best_loss"][:]
    iters_cma = f['iters'][:]+1


with h5py.File("../study_opt_local/results/results_LM_train_V4.hdf5", "r") as f:
    best_loss_LM = f["best_loss"][:]
    iters_LM = f['iters'][:]+1


with h5py.File("../study_opt_model/results/results_GP_train_V4.h5", "r") as f:
    best_loss_GP = f['best_loss'][:]
    iters_GP = f['iters'][:]+1

```

```python

###* heierarchical results

with h5py.File("../study_opt_hierarchical/results/results_hier0_train_V4.h5", "r") as f:
    best_loss_hier0 = f['best_loss'][:]
    iters_hier0 = f['iters'][:]


with h5py.File("../study_opt_hierarchical/results/results_hier1_train_V4.h5", "r") as f:
    best_loss_hier1 = f['best_loss'][:]
    iters_hier1 = f['iters'][:]


with h5py.File("../study_opt_hierarchical/results/results_hier2_train_V4.h5", "r") as f:
    best_loss_hier2 = f['best_loss'][:]
    iters_hier2 = f['iters'][:]


with h5py.File("../study_opt_hierarchical/results/results_hier3_train_V4.h5", "r") as f:
    best_loss_hier3 = f['best_loss'][:]
    iters_hier3 = f['iters'][:]


with h5py.File("../study_opt_hierarchical/results/results_hier4_train_V4.h5", "r") as f:
    best_loss_hier4 = f['best_loss'][:]
    iters_hier4 = f['iters'][:]

```

```python

fig, ax = plt.subplots(figsize=(12, 8))
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'

color_vec = ["blue", "orange", "green", "red", "purple", "brown", "pink"]

###* baselines
ax.plot(iters_de, best_loss_de, linewidth=3, label="DE", color=color_vec[0])
ax.plot(iters_cma, best_loss_cma, linewidth=3, label="CMA-ES", color=color_vec[1])
ax.plot(iters_powell, best_loss_powell, linewidth=3, label="Powell", color=color_vec[2])
ax.plot(iters_GP, best_loss_GP, linewidth=3, label="GP", color=color_vec[3])
ax.plot(iters_LM, best_loss_LM, linewidth=3, label="TRF", color=color_vec[4])

###* hierarchical
ax.plot(iters_hier0, best_loss_hier0, linewidth=3.5, label="Hierarchical Exact", color=color_vec[5])
ax.plot(iters_hier1, best_loss_hier1, linewidth=3.5, label="Hierarchical Stoch k=18", color=color_vec[6])
# ax.plot(iters_hier2, best_loss_hier2, linewidth=3.5, label="Hier Stoch k=10")
# ax.plot(iters_hier3, best_loss_hier3, linewidth=3.5, label="Hier Stoch k=5")
# ax.plot(iters_hier4, best_loss_hier4, linewidth=3.5, label="Hier Stoch k=3")

ax.axhline(y=0.1038, linewidth=3, linestyle='dotted', color='tan', label="Default")


plt.xscale('log')
plt.yscale('log')
plt.ylim(0.05, 3.0)
plt.xlabel(r'Iterations', fontsize=32)
plt.ylabel(r'$\Phi_{\rm train}$', fontsize=35)
ax.tick_params(axis='x', labelsize=32)
ax.tick_params(axis='y', labelsize=36)
ax.legend(fontsize=22)

plt.grid(axis='x', which='major', linestyle='-', linewidth=0.8, alpha=0.5)
ax.axhline(y=1.0, color='gray', linestyle='-', linewidth=0.8, alpha=0.5)


plt.tight_layout()
plt.savefig("panel_results_all_algos_train_V4.png")
```
